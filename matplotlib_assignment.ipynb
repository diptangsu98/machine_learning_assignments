# Matplotlib Assignment

## Scipy:

### We have the min and max temperatures in a city In India for each months of the year. We would like to find a function to describe this and show it graphically, the dataset given below.

## Data:

### Max=39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25

### Min=21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
max_temperature=np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min_temperature=np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
import matplotlib.pyplot as plt
months=np.arange(12)
plt.figure(figsize=(10,8))
plt.plot(months, max_temperature, 'ro')
plt.plot(months, min_temperature, 'bo')
plt.xlabel('Months')
plt.ylabel('Minimum and maximum temperature')
```

## Task:

### 1. Fitting it to the periodic function:

```python
from scipy import optimize
def yearly_temperatures(times, average, amplitude, time_offset):
    return (average+amplitude*np.cos((times+time_offset)*2*np.pi/times.max()))
res_maximum, cov_maximum=optimize.curve_fit(yearly_temperatures, months, max_temperature, [20, 10, 0])
res_minimum, cov_minimum=optimize.curve_fit(yearly_temperatures, months, min_temperature, [-40, 20, 0])
```

### 2. Plot the fit

```python
days=np.linspace(0, 12, num=365)
plt.figure(figsize=(10,8))
plt.plot(months, max_temperature, 'ro')
plt.plot(days, yearly_temperatures(days, *res_maximum), 'r-')
plt.plot(months, min_temperature, 'bo')
plt.plot(days, yearly_temperatures(days, *res_minimum), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')
plt.show()
```

## Matplotlib

### This assignment is for visualization using matplotlib:

### Data to use:

###  https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv

```python
import pandas as pd
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic=pd.read_csv(url)
```

## Charts to plot

### 1. Create a pie chart presenting the male/female proportion

```python
gender=titanic['sex'].astype('category')
gender=gender.cat.codes
```

```python
plt.pie(gender.value_counts(), labels=titanic['sex'].unique(), autopct='%1.1f%%')
plt.title('Male/Female Proportion')
plt.axis('equal')
plt.show()
```

### 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
plt.figure(figsize=(10,8))
plt.scatter(titanic['age'], titanic['fare'], c=gender)
plt.xlabel("Age")
plt.ylabel("Fare Paid")
plt.show()
```
