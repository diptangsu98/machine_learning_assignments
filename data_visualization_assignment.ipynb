## Data Visualization Assignment

### In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
i=datasets.load_iris()
a=i.data
b=i.target
```

```python
f=plt.figure(figsize=(10, 8))
plt.clf()
axes=Axes3D(f, rect=[0, 0, 0.9, 1], elev=50, azim=140)
plt.cla()
pca=decomposition.PCA(n_components=3)
pca.fit(a)
a=pca.transform(a)
b=np.choose(b, [1, 2, 0]).astype(np.float)
for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    axes.text3D(a[b==label, 0].mean(),
              a[b==label, 1].mean()+1.5,
              a[b==label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=0.4, edgecolor='b', facecolor='w'))
axes.scatter(a[:, 0], a[:, 1], a[:, 2], c=b, edgecolor='k')
plt.show()
```
